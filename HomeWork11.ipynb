```python

from matplotlib import pyplot as plt
from sympy import *
from sympy.plotting import plot

x = Symbol('x')
f= 0.6*x**3 + 5.5*x**2 + 10*x -5

lst = list(solveset(f,x))

print("Корни:")
lst = [print(f'x{i+1} = {lst[i]}') for i in range(len(lst))]
```

```python
from sympy import *
# нахождение производной и областей возрастания, убывания

x = Symbol('x')
y =  0.6*x**3 + 5.5*x**2 + 10*x -5
yprime = y.diff(x)
print(yprime) 
sp = solve(yprime)
print(sp)

print("\n")
print("Функция возрастает в области от (-оо, -5), (-1,111111, оо)")
print('Функция убывает (-5,-1.111111)')
```

```python
# Находим Вершины

# при x = -5
y1 =  0.6*(-5)**3 + 5.5*(-5)**2 + 10*(-5) -5
print(f"Перва вершина ({-5}, {y1})")

# при x = -1.11111111111111
y2 = 0.6*(-1.11111111111111)**3 + 5.5*(-1.11111111111111)**2 + 10*(-1.11111111111111) -5
print(f"Вторая вершина ({-1.11111111111111}, {y2})")
```
